```python
import pandas as pd, matplotlib.pyplot as plt, seaborn as sns
```

```python
# Load cleaned data
df = pd.read_csv("titanic_cleaned.csv")
# Quick check
df.head()
```

```python
# Set style
sns.set_style("whitegrid")
plt.figure(figsize=(20,14))


# Row 1, col 1: Survival by Sex
plt.subplot(2,3,1)
sns.countplot(x = "Sex", hue = "Survived", data = df)
plt.title("1. Survival by Sex")

# Row 1, Col 2: Survival by Pclass
plt.subplot(2, 3, 2)
sns.countplot(x = "Pclass", hue = "Survived", data = df)
plt.title("2. Survival by Passenger Class")

plt.subplot(2, 3, 3)
sns.histplot(data = df, x = "Age", hue = "Survived", bins = 30, kde = True)
plt.title("3. Age Distribution by Survival")


plt.subplot(2, 3, 4)
sns.boxplot(x = "Survived", y = "Fare",  data = df)
plt.title("4. Fare by Survival")

plt.subplot(2, 3, 5)
sns.countplot(x = "SibSp", hue = "Survived", data = df)
plt.title("5. Family Size by Survival")

plt.subplot(2, 3, 6)
sns.countplot(x = "Embarked", hue = "Survived", data = df)
plt.title("6. Survival by Embarkation Port")

plt.tight_layout()
plt.show()
```

```python
sns.catplot(
    x = "Pclass",
    hue = "Survived",
    col = "Sex",
    data = df,
    kind = "count",
    height = 6,
)
plt.subplots_adjust(top = 0.85)
plt.suptitle("7. Survival by Pclass and Sex", fontsize = 16)
plt.show()
```
